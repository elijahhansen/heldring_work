```python

import re

def scrape_and_filter(file_path, stopwords):
    with open(file_path, 'r') as file:
        content = file.read()

    # Convert stopwords to a set for faster membership lookup
    stopwords = set(stopwords)

    # Split the content into words
    words = re.findall(r'\b\w+\b', content)

    # Filter out stopwords and lowercase the words
    filtered_words = [word.lower() for word in words if word.lower() not in stopwords]

    return filtered_words

# Example usage
file_path = 'Locke.txt'
stopwords = ['the', 'and', 'to', 'is', 'in', 'it', 'a', 'an','of','in','that']

results = scrape_and_filter(file_path, stopwords)
print(results)
```

```python

import re
from collections import Counter

def find_highly_referenced_words(file_path, num_words):
    with open(file_path, 'r') as file:
        content = file.read()

    # Split the content into words
    words = re.findall(r'\b\w+\b', content)

    # Count the frequency of each word
    word_counts = Counter(words)

    # Get the most frequent words
    highly_referenced_words = word_counts.most_common(num_words)

    return highly_referenced_words

# Example usage
file_path = 'Filmer.txt'
num_words = 500

results = find_highly_referenced_words(file_path, num_words)
print(results)

```

```python
import re
from collections import Counter
from nltk import ngrams

def find_highly_referenced_expressions(file_path, num_expressions, expression_length):
    with open(file_path, 'r') as file:
        content = file.read()

    # Split the content into words
    words = re.findall(r'\b\w+\b', content)

    # Generate n-grams of the specified length
    expressions = ngrams(words, expression_length)

    # Count the frequency of each expression
    expression_counts = Counter(expressions)

    # Get the most frequent expressions
    highly_referenced_expressions = expression_counts.most_common(num_expressions)

    return highly_referenced_expressions

# Example usage
file_path = 'Filmer.txt'
num_expressions = 100
expression_length = 3

results = find_highly_referenced_expressions(file_path, num_expressions, expression_length)
print(results)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(content)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Example usage:
input_text = 'Locke.txt'
result = remove_stopwords(input_text)
print(result)
```

```python
import re
from nltk.tokenize import sent_tokenize

def find_sentences_with_expressions(file_path, expressions):
    """
    extracts sentences that use expressions from inputted list
    ---
    file_path: string
    expressions: list[tuple]
    """
    with open(file_path, 'r') as file:
        content = file.read()

    # Tokenize the content into sentences
    sentences = sent_tokenize(content)

    # Find sentences that contain the given expressions
    sentences_with_expressions = []
    for sentence in sentences:
        for expr in expressions:
            if all(word.lower() in sentence.lower() for word in expr[0]):
                sentences_with_expressions.append(sentence)

    return sentences_with_expressions

# Example usage
file_path_locke = 'Locke.txt'
file_path_filmer = 'Filmer.txt'
expressions = 'find_most_common_ngrams_list'

sentences_with_expressions_locke = find_sentences_with_expressions(file_path_locke, expressions)

```

# Ngram Extraction (Frequency)

```python
from typing_extensions import Concatenate
import nltk
from nltk.util import ngrams
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

nltk.download('punkt')
nltk.download('stopwords')

def get_most_common_ngrams(file_path, n, num_most_common):
    """
    file_path: str
    n: int; length of desired ngrams
    top_n: int; number of desired ngrams
    """
    with open(file_path, 'r') as file:
        content = file.read()
    words = word_tokenize(content)
    stop_words = set(stopwords.words('english') + list(string.punctuation))
    words = [word for word in words if word.lower() not in stop_words]
    n_grams = ngrams(words, n)
    n_gram_freq = Counter(n_grams)
    return n_gram_freq.most_common(num_most_common)

# Example usage:
input_text = 'Locke.txt'
n_value = 3  # Change this value to the desired 'n' for n-grams (e.g., 2 for bigrams, 3 for trigrams, etc.)
num_most_common_ngrams = 100  # Change this value to get more or fewer results
result = get_most_common_ngrams(input_text, n_value, num_most_common_ngrams)
print(result)
```

```python
import nltk
from collections import Counter

nltk.download('stopwords')

def find_most_common_ngrams_list(file_path, n=2, top_n=10):
    """
    same functionality as "get_most_common_ngrams()" but returns a list of str
    instead of a list of tuples
    ---
    file_path: str
    n: int; length of desired ngrams
    top_n: int; number of desired ngrams
    """
    with open(file_path, 'r') as file:
        content = file.read()
    text = content
    # Tokenize the text into words
    words = nltk.word_tokenize(text)

    # Generate n-grams
    ngrams_list = [' '.join(words[i:i + n]) for i in range(len(words) - n + 1)]

    # Filter out n-grams containing stopwords
    stopwords = set(nltk.corpus.stopwords.words('english'))
    ngrams_list = [ngram for ngram in ngrams_list if not any(word in stopwords for word in ngram.split())]

    # Count the occurrences of each n-gram
    ngrams_counter = Counter(ngrams_list)

    # Find the top_n most common n-grams
    most_common_ngrams = ngrams_counter.most_common(top_n)

    # Extract only the n-grams without frequencies
    most_common_ngrams = [ngram for ngram, _ in most_common_ngrams]

    return most_common_ngrams
```

# TFIDF Ngram Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def find_unique_ngrams_tfidf(text1, text2, n=3, top_n=100):
    """
    find n-grams in text1 based on inverse frequency of n-grams in text2
    ---
    text1: str
    text2: str
    n: int; length of desired ngrams
    top_n: int; number of desired ngrams
    """
    # Combine the texts into a list of two sentences
    sentences = [text1, text2]

    # Initialize TfidfVectorizer with the desired n-gram range
    vectorizer = TfidfVectorizer(ngram_range=(n, n))

    # Calculate TF-IDF scores for both texts
    tfidf_scores = vectorizer.fit_transform(sentences)

    # Get feature names (n-grams) and their corresponding TF-IDF scores for both texts
    feature_names = vectorizer.get_feature_names_out()
    tfidf_scores_text1 = tfidf_scores.toarray()[0]
    tfidf_scores_text2 = tfidf_scores.toarray()[1]

    # Combine feature names and TF-IDF scores into a list of tuples for both texts
    ngrams_with_scores_text1 = list(zip(feature_names, tfidf_scores_text1))
    ngrams_with_scores_text2 = list(zip(feature_names, tfidf_scores_text2))

    # Sort n-grams in both texts based on TF-IDF scores in descending order
    sorted_ngrams_text1 = sorted(ngrams_with_scores_text1, key=lambda x: x[1], reverse=True)
    sorted_ngrams_text2 = sorted(ngrams_with_scores_text2, key=lambda x: x[1], reverse=True)

    # Extract top_n n-grams with highest TF-IDF scores for both texts
    top_ngrams_text1 = [ngram for ngram, _ in sorted_ngrams_text1[:top_n]]
    top_ngrams_text2 = [ngram for ngram, _ in sorted_ngrams_text2[:top_n]]

    # Find n-grams used in text1 that are not used in text2
    unique_ngrams_text1 = [ngram for ngram in top_ngrams_text1 if ngram not in top_ngrams_text2]

    return unique_ngrams_text1

# Example usage
file_path1 = 'Locke.txt'
with open(file_path1, 'r') as file:
        content1 = file.read()
text1 = content1
file_path2 = 'hamlet.txt'
with open(file_path2, 'r') as file:
        content2 = file.read()
text2 = content2
n = 4
top_n = 100

#Unique list of ngrams
unique_ngrams = find_unique_ngrams_tfidf(text1, text2, n, top_n)
extracted_sentences = find_sentences_with_expressions(file_path1,unique_ngrams)
#Unique Sentences
unique_sentences_tfidf = find_unique_items(extracted_sentences)
print(len(unique_sentences_tfidf))


print(f"{n}-grams used in text1 that are not used in text2:")
for ngram in unique_ngrams:
    print(f"N-gram: '{ngram}'")
```

#Cosine Similarity Calculations

```python
import nltk
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')

def calculate_average_cosine_similarity(sentences_list, text, score_threshold):
    """
    calculates cosine similarity scores for a list of sentences and a text then
    selects the sentences with a score higher than the score_threshold. it does
    this for both the average scores and the max scores between the combinations
    of each element with the text

    sentences_list: list[str]
    text: str
    score_threshold: float; designates the cutoff score for cosine similarity
    """
    # Tokenize sentences in the list and the text
    sentences_list = [nltk.sent_tokenize(sentence) for sentence in sentences_list]
    text_sentences = nltk.sent_tokenize(text)

    # Flatten the list of sentences
    sentences_list = [sentence for sublist in sentences_list for sentence in sublist]

    # Create the SentenceTransformer model
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')


    # Encode sentences into embeddings
    embeddings_list = model.encode(sentences_list, show_progress_bar=True)
    embeddings_text = model.encode(text_sentences, show_progress_bar=True)


    # Calculate cosine similarity between the embeddings of the list and the text
    similarity_scores = cosine_similarity(embeddings_list, embeddings_text)




    # Calculate average cosine similarity for each sentence in the list
    average_similarities = np.mean(similarity_scores, axis=0)
    max_similarities = np.max(similarity_scores, axis=0)
    argmax_avg = np.argmax(average_similarities)
    argmax_max = np.argmax(max_similarities)
    print(average_similarities.shape, max_similarities.shape)


    # Combine sentences and their average similarity scores into a list of tuples

    sentences_outside_avg = list(zip(text_sentences,average_similarities))
    sentences_outside_max = list(zip(text_sentences,max_similarities))

    #filter sentences by score threshold
    filtered_sentences_avg = [(sentence,score) for sentence, score in sentences_outside_avg if score >= score_threshold]
    filtered_sentences_max = [(sentence,score) for sentence, score in sentences_outside_max if score >= score_threshold]

    # Sort sentences based on average similarity scores in descending order
    sorted_sentences_avg = sorted(filtered_sentences_avg, key=lambda x: x[1], reverse=True)
    sorted_sentences_max = sorted(filtered_sentences_max, key=lambda x: x[1], reverse=True)

    return sorted_sentences_avg, sorted_sentences_max

# Example usage
file_path = 'Filmer.txt'
with open(file_path, 'r') as file:
  content = file.read()
text = content
#find (num_grams) ngrams of length (ngram_length) in text
ngram_length = 3
num_ngrams = 100
expressions = get_most_common_ngrams(file_path, 3, 100)
#extract unique sentences from text that use (expressions)
sentences_with_expressions = find_sentences_with_expressions(file_path, expressions)
sentences_list = find_unique_items(sentences_with_expressions)
#find similar sentences using function defined above
score_threshold = 0.430
similar_sentences = calculate_average_cosine_similarity(sentences_list, text, score_threshold)

print("Sentences with the highest cosine similarity to the text:")
print(len(similar_sentences[1]))
for sentence, similarity_score in similar_sentences[1]:
    print(f"Sentence: '{sentence}', Similarity Score: {similarity_score:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_excel('titles_with_dates.xlsx')

# Split the "entries" column into a list of classifications using commas
df['classifications'] = df['entries'].str.split(', ')

# Calculate the total number of entries (before exploding)
total_entries = df['entries'].count()

# Explode the list of classifications into separate rows
df = df.explode('classifications')

# Group the data by the "session" column and calculate rates for each session
session_grouped = df.groupby(['session', 'classifications']).size().unstack(fill_value=0)
print(session_grouped)


# Calculate the rate as a percentage of total entries in each session
session_grouped = (session_grouped / total_entries) * 100

# Plot the rates for each session
session_grouped.plot(kind='line', figsize=(10, 6))
print(session_grouped)

# Customize the plot
plt.xlabel('Session')
plt.ylabel('Rate (%)')
plt.title('Rate of Each Classification by Session')
plt.legend(title='Classification', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
